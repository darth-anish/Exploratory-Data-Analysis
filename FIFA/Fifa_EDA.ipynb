# Data Preprocessing

```python
#removing € with nothing
fifa_df['Value']=fifa_df['Value'].str.replace('€','')
```

```python
#converting M and k into float
fifa_df['Value']=fifa_df['Value'].replace({'K': '*1e3', 'M': '*1e6'}, regex=True).map(pd.eval).astype(int)
```

```python
fifa_df
```

```python
#converting wage and value 
```

```python
fifa_df['Wage']=fifa_df['Wage'].str.replace('€','')
```

```python
fifa_df['Wage']=fifa_df['Wage'].replace({'K':'*1e3','M':'*1e6'},regex=True).map(pd.eval).astype(int)
```

```python
fifa_df['Wage']=fifa_df['Wage'].astype(float)
```

```python
fifa_df['Value']=fifa_df['Value'].astype(float)
```

```python
fifa_df['Value'].replace(0,fifa_df['Value'].median(),inplace=True)
```

```python
fifa_df['Value'].value_counts()
```

```python
fifa_df['Wage'].replace(0,fifa_df['Wage'].median(),inplace=True)
```

```python
fifa_df['Wage'].value_counts()
```

```python
#making a column named joined_year 

fifa_df['joined_year']=fifa_df['Joined'].str.split(',').str[1]
```

```python
fifa_df['joined_year']=fifa_df['joined_year'].fillna(0).astype(int)
```

```python
fifa_df['joined_year']
```

```python
#changing a string to datetime format 

fifa_df['Contract Valid Until']=pd.to_datetime(fifa_df['Contract Valid Until'])
```

```python
fifa_df['Contract Valid Until']
```

```python
fifa_df['Height']
```

```python
fifa_df['Height']=fifa_df['Height'].str.split("'").str[0].astype(float)+(fifa_df['Height'].str.split("'").str[1].astype(float))*0.1
```

```python
fifa_df['Height']=np.where(fifa_df['Height']==np.nan,fifa_df['Height'].median(),fifa_df['Height'])
```

```python
fifa_df['Height'].fillna(fifa_df['Height'].median(),inplace=True)
```

```python
fifa_df['Height'].isnull().sum()
```

```python
fifa_df
```

```python
fifa_df['Weight']=fifa_df['Weight'].str.strip(to_strip='lbs')
```

```python
fifa_df['Weight'].astype(float)
```

### Converting release clause column into float 

```python
fifa_df['Release Clause']=fifa_df['Release Clause'].str.replace('€','')
```

```python
fifa_df['Release Clause'].isnull().sum()
```

```python
fifa_df['Release Clause']=fifa_df['Release Clause'].fillna(0)
```

```python
fifa_df['Release Clause']=fifa_df['Release Clause'].replace({'K':'*1e3','M':'*1e6'},regex=True).map(pd.eval).astype(float)
```

```python
fifa_df['Release Clause']=np.where(fifa_df['Release Clause']==0,fifa_df['Release Clause'].median(),fifa_df['Release Clause'])
```

```python
fifa_df['Release Clause'].isnull().sum()
```

# Exploratory Data Analysis

## Right Vs Left Footed count plot

```python
pf_grp=fifa_df.groupby('Preferred Foot')
pf_grp['Preferred Foot'].count().plot(kind='bar')
plt.ylabel('No of players')
plt.show()
```

## Release Clause of top 5 players

```python
rc_5=fifa_df[['Name','Release Clause']].head(5)

fig, ax = plt.subplots(figsize=(8,5))

sns.barplot(data=rc_5,x='Name',y='Release Clause',ax=ax)

ax.set(ylabel='Release Clause in £',xlabel='Player Name')

plt.show()
```

## Pie chart of top 5 football producing countries

```python
val_nat=fifa_df['Nationality'].value_counts().head()  

plt.figure(figsize=(8,5))

plt.pie(val_nat.values,labels=val_nat.index, autopct='%1.1f%%')

plt.show()
```

## Top 5 Wage Paying Clubs

```python
f_n=fifa_df.groupby('Club')

club_wg_df=pd.DataFrame()

club_wg_df=pd.DataFrame(f_n['Wage'].mean())

club_wg_df.reset_index(level=0,inplace=True)
club_wg_df['Rank_wage']=club_wg_df['Wage'].rank(ascending=False)
club_wg_df.sort_values('Rank_wage',inplace=True)
```

```python
club_wg_df.head()
```

```python
wg_5=club_wg_df[['Club','Wage']].head(5)

fig, ax = plt.subplots(figsize=(8,5))

sns.barplot(data=wg_5,x='Club',y='Wage',ax=ax)

ax.set(ylabel='Average Wage',xlabel='Club Name')

plt.show()
```
